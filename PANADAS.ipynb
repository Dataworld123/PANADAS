Q4. What is ‘DataFrame’ in pandas and how is it different from pandas.series? Explain with an example.

ANS- In Pandas, a DataFrame is a two-dimensional labeled data structure that can store data of different types (numeric, string, boolean, etc.) in a tabular format. It is similar to a spreadsheet or SQL table, where data is organized in rows and columns. Each column in a DataFrame represents a variable, and each row represents an observation or record.

On the other hand, a Pandas Series is a one-dimensional labeled array that can hold data of a single data type, like a list or an array. It is similar to a column in a DataFrame, but it doesn't have column names. Instead, it has an index, which labels each element in the Series.

Here's an example to illustrate the difference between a DataFrame and a Series:

import pandas as pd

# Example 1: Creating a Pandas Series
fruits = pd.Series(['apple', 'banana', 'orange', 'grape', 'watermelon'])
print("Pandas Series:")
print(fruits)


# Example 2: Creating a Pandas DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Claire'],
    'Age': [25, 30, 27],
    'Gender': ['Female', 'Male', 'Female']
}

df = pd.DataFrame(data)
print("\nPandas DataFrame:")
print(df)


Q5. What are some common functions you can use to manipulate data in a Pandas DataFrame? Can
you give an example of when you might use one of these functions?


ANS- Pandas provides a wide range of functions to manipulate data in a DataFrame. Some common functions include:

head() and tail(): These functions allow you to view the first few rows (head()) or the last few rows (tail()) of the DataFrame. They are useful for quickly inspecting the data to get a sense of its structure.

import pandas as pd

# Create a DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Claire', 'David', 'Eva'],
    'Age': [25, 30, 27, 22, 29],
    'Gender': ['Female', 'Male', 'Female', 'Male', 'Female']
}

df = pd.DataFrame(data)

# Use head() to view the first 3 rows
print(df.head(3))

# Use tail() to view the last 2 rows
print(df.tail(2))


Q6. Which of the following is mutable in nature Series, DataFrame, Panel?

ANS-In Pandas, both Series and DataFrame are mutable in nature, while Panel is not.

Series: A Pandas Series is a one-dimensional labeled array, and it is mutable. You can modify the elements of a Series after it has been created.

DataFrame: A Pandas DataFrame is a two-dimensional labeled data structure, and it is also mutable. You can modify the data in the DataFrame, including adding or removing rows and columns, updating values, etc.

Panel: A Panel is a three-dimensional data structure in Pandas, which was used to handle 3D data in older versions. However, it has been deprecated, and the recommendation is to use other data structures like Series and DataFrame for most use cases. Panels are not mutable like Series and DataFrame.

As of Pandas version 1.0.0, Panels are considered deprecated, and you might encounter a recommendation to use other data structures like MultiIndex DataFrame or xarray to handle three-dimensional data.

So, to summarize, Series and DataFrame are both mutable in nature, but Panel is not and is considered deprecated.

Q7. Create a DataFrame using multiple Series. Explain with an example.

ANS-Creating a DataFrame using multiple Series is a common operation in Pandas. You can combine several Series with different lengths, data types, or sources to form a DataFrame, where each Series represents a column in the resulting DataFrame.

Let's create a DataFrame using two example Series: one for student names and another for their corresponding ages.

import pandas as pd

# Create the first Series for student names
names = pd.Series(['Alice', 'Bob', 'Claire', 'David', 'Eva'])

# Create the second Series for student ages
ages = pd.Series([25, 30, 27, 22, 29])

# Combine the Series to create the DataFrame
df = pd.DataFrame({'Name': names, 'Age': ages})

# Print the DataFrame
print(df)
